```python
import requests
import random


# Generar datos de prueba según el schema
def generate_test_data(num_samples=5):
    # Lista de posibles códigos de mercancía
    productos = [
        "Café Tostado", "Azúcar Refinado", "Arroz Blanco", "Aceite Vegetal",
        "Harina de Trigo", "Leche en Polvo", "Cacao en Grano", "Sal Refinada",
        "Frijol Negro", "Maíz Amarillo"
    ]
    test_data = []
    for _ in range(num_samples):
        registro = {
            "codigo_mercancia": random.choice(productos),
            "cnt_merc": round(random.uniform(50, 1000), 2),
            "fob_unitario": round(random.uniform(1.5, 15.0), 2)
        }
        test_data.append(registro)
    return test_data


# Generar 5 muestras de prueba
df_request = generate_test_data(5)

try:
    for i in range(len(df_request)):
        response = requests.post(
            "http://20.29.159.59:8000/v1",
            json=df_request[i]
        )
        result = response.json()
        print(df_request[i], "--> ", result)
        print("-" * 120)

except Exception as ex:
    print(f"Request: Se produjo el siguiente error : {ex} ")
```
